```python
import csv

max_temp = []
min_temp = []

with open('seoul.csv', 'r', encoding='cp949') as f:
  data = csv.reader(f)
  header = next(data)

  for row in data:
    if row[3] == '':
      row[3] = 999
    if row[4] == '':
      row[4] = -999

    row[3] = float(row[3])
    row[4] = float(row[4])

    min_temp.append(row[3])
    max_temp.append(row[4])

print(f'서울 최고 기온 10개')
max_temp.sort(reverse=True)
print(max_temp[:10])

print(f'서울 최저 기온 10개')
min_temp.sort()
print(min_temp[:10])
```
